```python
# Evaluation and Optimization of Mass Balance Calculation Methods  
## Forced Degradation Studies – Novartis NEST 2025

### Objective
To develop a structured, uncertainty-aware framework for interpreting mass balance
results in forced degradation studies, reducing subjective judgment and unnecessary investigations.


## 1. Forced Degradation Studies 

Forced degradation studies are performed to intentionally degrade a drug substance
using stress conditions such as acid, base, heat, oxidation, and light.

The purpose is to:
- Understand degradation pathways
- Ensure the analytical method can detect degradation products
- Prove the method is stability-indicating

In simple words:
We purposely damage the drug to see how it breaks and whether our method can detect everything.
```

```python
## 2. Mass Balance (MB)

Mass Balance checks whether the total amount of drug-related material after stress
(API remaining + degradation products) accounts for the initial drug amount.

Initial drug amount = 100%
After stress:
API remaining (%) + Degradants (%) ≈ 100%

### Why Mass Balance Matters

Regulatory agencies expect mass balance values to remain close to 100% because this indicates:
- All degradation products are detected
- Minimal physical loss (volatility, adsorption)
- Adequate analytical method sensitivity

If this is true → method is reliable
If not → something is missing or wrong
```

```python
## 3. Problem Statement

In practice, mass balance interpretation is inconsistent and subjective because:
- Multiple mass balance calculation methods exist
- Fixed acceptance thresholds are used
- Analytical uncertainty is not explicitly considered

The same mass balance value can lead to different conclusions by different analysts.

100% ≠ perfection
Close to 100% = acceptable
```

```python
## 4. Existing Mass Balance Metrics

| Metric | Definition         | Limitation                  |
| ------ | ------------------ | --------------------------- |
| SMB    | API + degradants   | No uncertainty context      |
| AMB    | Absolute recovery  | Sensitive to noise          |
| RMB    | Normalized MB      | Can mask missing degradants |
| AMBD   | Deviation from 100 | No diagnosis                |
| RMBD   | RMB deviation      | Still single-value          |
```

```python
## 5. Method Variability

All analytical methods have inherent variability due to:
- Instrument noise
- Sample preparation
- Injection and integration differences

Method variability represents the expected analytical uncertainty of a validated method.

Example:

Method variability = ±2%
MB between 98% and 102% can be normal
Deviation ≠ problem automatically
```

```python
## 6. Uncertainty-Adjusted Metric (Z_MB)

To distinguish real mass loss from analytical noise, an uncertainty-adjusted parameter
called Z_MB is introduced.

Z_MB = (AMB − 100) / Method Variability

Interpretation:

Small |Z_MB| → normal analytical noise

Large |Z_MB| → real concern

Z_MB answers:

“Is this deviation significant compared to expected error?”
```

```python
## 7. Proposed Mass Balance Diagnostic Framework

The framework integrates:
- Absolute Mass Balance (AMB)
- Relative Mass Balance (RMB)
- Method Variability (Z_MB)

Instead of relying on a single value, results are classified into diagnostic zones.

Raw Data → MB Metrics → Z_MB → Diagnostic Zone → Recommendation
```

```python
## 8. Diagnostic Zones

| Zone   | Meaning                       |
| ------ | ----------------------------- |
| Zone 1 | Within analytical variability |
| Zone 2 | Possible missing degradants   |
| Zone 3 | Likely physical loss          |
| Zone 4 | Possible overestimation       |

This converts numbers → diagnosis
```

```python
## 9. Interpretation

Interpretation explains the scientific meaning of a mass balance result.
It answers:
“Is the deviation analytically explainable or scientifically concerning?”

Interpretation:
Mass balance deviation is within method variability and does not indicate true loss.
```

```python
## 10. Recommendation Matrix

| Zone   | Interpretation     | Recommendation    |
| ------ | ------------------ | ----------------- |
| Zone 1 | Normal variability | Accept            |
| Zone 2 | Missing degradants | Review method     |
| Zone 3 | Physical loss      | Investigate       |
| Zone 4 | Overestimation     | Check calibration |
```

```python
## 11. Comparison with Traditional Approach

| Aspect                   | Traditional MB | Proposed Framework |
| ------------------------ | -------------- | ------------------ |
| Variability considered   | No             | Yes                |
| Explains cause           | No             | Yes                |
| Decision consistency     | Low            | High               |
| Investigation efficiency | Poor           | Improved           |
```
