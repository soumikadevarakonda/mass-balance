```python
# Mass Balance Diagnostic Framework  
### Forced Degradation Studies – Novartis NEST 2025  
**Team BitByBoth**

This notebook provides a technical walkthrough of the Mass Balance Diagnostic Framework proposed for improving interpretation of mass balance results in forced degradation studies.

The objective is not to introduce a new mass balance formula, but to demonstrate how existing mass balance metrics can be jointly interpreted using analytical uncertainty and degradation behavior to support consistent, risk-based decision-making.
```

```python
## 1. Why Mass Balance Interpretation Fails in Practice

Mass balance (MB) is expected to remain close to 100% in forced degradation studies, indicating that API loss is adequately explained by detected degradation products.

In practice, interpretation is inconsistent because:
- Multiple mass balance calculation methods coexist
- Fixed numerical thresholds are applied without uncertainty context
- Analytical variability is not explicitly incorporated
- Similar mass balance values may arise from fundamentally different causes

As a result, analysts often face unnecessary investigations or unclear root-cause attribution despite comparable numerical outcomes.
```

```python
## 2. Existing Mass Balance Metrics (Summary)

| Metric | Purpose | Key Limitation |
|------|--------|---------------|
| AMB | Total recovery of API + degradants | Sensitive to analytical noise |
| RMB | Relates degradants to API loss | Unstable at low degradation |
| AMBD | Magnitude of deviation from 100% | No diagnostic insight |
| RMBD | Deviation of RMB from unity | Still single-value interpretation |

These metrics quantify deviation but do not explain *why* it occurs.
```

```python
## 3. Role of Analytical Variability (σ)

All validated analytical methods exhibit inherent variability due to:
- Instrument precision
- Sample preparation effects
- Injection, integration, and analyst differences

In this framework, analytical variability (σ) is derived from **intermediate precision data** generated during method validation, in accordance with ICH Q2(R1).

σ represents the expected spread of analytically acceptable results and forms the statistical baseline for interpreting mass balance deviation.
```

```python
## 4. Uncertainty-Adjusted Mass Balance Index (Z_MB)

To contextualize mass balance deviation, an uncertainty-adjusted index is defined:

Z_MB = (AMB − 100) / σ

Interpretation:
- |Z_MB| ≤ 2 → Deviation within ~95% confidence (analytically explainable)
- |Z_MB| ≥ 3 → Deviation beyond ~99% confidence (likely systematic)

This converts mass balance from a raw percentage into a statistically interpretable signal.
```

```python
## 5. Why Z_MB Alone Is Not Sufficient

Z_MB quantifies the *significance* of deviation but does not explain the *mechanism*.

For example:
- Low AMB may arise from missing degradants or physical loss
- High AMB may arise from response factor mismatch or integration bias

Therefore, deviation magnitude must be interpreted jointly with recovery behavior to enable root-cause diagnosis.
```

```python
## 6. AMB–RMB Diagnostic Map (Core Diagnostic Layer)

Joint interpretation of Absolute Mass Balance (AMB) and Relative Mass Balance (RMB) enables mechanistic classification:

| AMB | RMB | Likely Cause |
|----|----|-------------|
| Normal | Normal | Consistent analytical recovery |
| Low | Low | Missing or non-chromophoric degradants |
| Low | Normal | Physical loss (adsorption, volatility) |
| High | High | Analytical overestimation (RRF or integration bias) |

This step converts numerical deviation into scientific diagnosis.
```

```python
## 7. Recommendation Matrix (Action-Oriented)

Each diagnostic zone maps to targeted analytical actions:

- Missing degradants → Orthogonal testing (LC–MS, CAD/ELSD, alternate column chemistry)
- Physical loss → Adsorption studies, headspace analysis for volatile degradants
- Overestimation → Verification of relative response factors (RRF) and integration review

This replaces generic investigations with focused, mechanism-driven actions.
```

```python
## 8. Importance of Trend Consistency

Consistent directional bias in Z_MB across stress conditions or time points suggests:
- Systematic response factor mismatch
- Integration bias
- Stability-indicating method failure

Random scatter within uncertainty limits indicates normal analytical behavior.

Thus, **mass balance stability over time** is as important as absolute mass balance value.
```

```python
## 9. Comparison with Traditional Mass Balance Interpretation

| Aspect | Traditional MB | Diagnostic Framework |
|------|----------------|---------------------|
| Analytical uncertainty considered | No | Yes |
| Root-cause insight | No | Yes |
| Investigation triggers | Fixed thresholds | Risk-based |
| Analyst dependence | High | Reduced |
```

```python
## 10. Key Takeaway

There is no single “best” mass balance formula.

Reliable interpretation emerges from:
- Statistical significance (Z_MB)
- Recovery behavior (AMB–RMB)
- Mechanistic diagnosis
- Actionable recommendations

This framework transforms mass balance from a compliance check into a scientifically defensible decision-support tool.
```
