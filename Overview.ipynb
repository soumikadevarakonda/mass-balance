# Evaluation and Optimization of Mass Balance Calculation Methods  
## Forced Degradation Studies – Novartis NEST 2025

**Team BitByBoth**

### Objective
To develop a structured, uncertainty-aware framework for interpreting mass balance results in forced degradation studies, reducing subjective judgment and unnecessary investigations.

## 1. Forced Degradation Studies

Forced degradation studies are performed to intentionally degrade a drug substance using stress conditions such as acid, base, heat, oxidation, and light.

The purpose is to:
- Understand degradation pathways
- Ensure the analytical method can detect degradation products
- Demonstrate that the method is stability-indicating

**In simple terms:** we intentionally stress the drug to observe how it degrades and whether the analytical method can detect all relevant degradation products.

## 2. Mass Balance (MB)

Mass balance evaluates whether the total drug-related material after stress accounts for the initial drug amount.

**Initial drug amount:** 100%  
**After stress:**

API remaining (%) + Degradation products (%) ≈ 100%

### Why Mass Balance Matters

Regulatory agencies expect mass balance values to remain reasonably close to 100% because this indicates:
- Adequate detection of degradation products
- Minimal physical loss (e.g., volatility or adsorption)
- Sufficient analytical method sensitivity

If mass balance is acceptable → the method is reliable.  
If not → something is missing or incorrectly measured.

## 3. Problem Statement

In practice, mass balance interpretation is inconsistent and subjective because:
- Multiple mass balance calculation methods coexist
- Fixed acceptance thresholds are commonly applied
- Analytical uncertainty is rarely considered explicitly

As a result, the same mass balance value may lead to different conclusions by different analysts.

**100% does not imply perfection.**  
**Close to 100% may still be scientifically acceptable.**

## 4. Existing Mass Balance Metrics

| Metric | Definition | Limitation |
|------|------------|------------|
| SMB | API + degradants | No uncertainty context |
| AMB | Absolute recovery | Sensitive to analytical noise |
| RMB | Normalized mass balance | Can mask missing degradants |
| AMBD | Deviation from 100% | No diagnostic insight |
| RMBD | RMB deviation | Still a single-value metric |

These metrics quantify deviation but do not explain its cause.

## 5. Method Variability

All analytical methods exhibit inherent variability due to:
- Instrument precision
- Sample preparation effects
- Injection and integration differences

Method variability represents the expected analytical uncertainty of a validated method.

**Example:**  
If method variability is ±2%, mass balance values between 98% and 102% may be analytically acceptable.

Deviation does not automatically imply a scientific problem.

## 6. Uncertainty-Adjusted Metric (Z_MB)

To distinguish real mass imbalance from analytical noise, an uncertainty-adjusted index is defined:

**Z_MB = (AMB − 100) / σ**

where σ represents analytical variability derived from method precision data.

**Interpretation:**
- Small |Z_MB| → analytically explainable variability
- Large |Z_MB| → statistically significant deviation

Z_MB answers the key question:

*“Is this deviation significant compared to expected analytical error?”*

## 7. Proposed Mass Balance Diagnostic Framework

The proposed framework integrates:
- Absolute Mass Balance (AMB)
- Relative Mass Balance (RMB)
- Analytical uncertainty (Z_MB)

Instead of relying on a single numerical value, results are interpreted through diagnostic classification.

**Workflow:**  
Raw Data → MB Metrics → Z_MB → Diagnostic Zone → Recommendation

## 8. Diagnostic Zones

| Zone | Meaning |
|------|---------|
| Zone 1 | Within analytical variability |
| Zone 2 | Possible missing degradants |
| Zone 3 | Likely physical loss |
| Zone 4 | Possible analytical overestimation |

This step converts numerical mass balance values into scientifically meaningful diagnoses.

## 9. Interpretation

Interpretation explains the scientific meaning of a mass balance result.

It answers:

*“Is the observed deviation analytically explainable or scientifically concerning?”*

**Example:**  
A deviation within method variability does not indicate true mass loss.

## 10. Recommendation Matrix

| Zone | Interpretation | Recommendation |
|------|---------------|----------------|
| Zone 1 | Normal variability | Accept result |
| Zone 2 | Missing degradants | Review method using orthogonal techniques |
| Zone 3 | Physical loss | Investigate adsorption or volatility |
| Zone 4 | Overestimation | Verify response factors and integration |

Each diagnostic outcome maps directly to a focused, mechanism-driven action.

## 11. Comparison with Traditional Approach

| Aspect | Traditional MB | Proposed Framework |
|-------|---------------|--------------------|
| Analytical variability considered | No | Yes |
| Root-cause insight | No | Yes |
| Decision consistency | Low | High |
| Investigation efficiency | Poor | Improved |


## 12. Conclusion

This work presents a structured and practical approach for optimizing mass balance evaluation in forced degradation studies.

By embedding analytical uncertainty and degradation behavior into a unified diagnostic framework, mass balance interpretation is transformed from a simple compliance metric into a decision-support tool.

**Mass balance deviations are inevitable — incorrect interpretations are not.**
